```python
import pandas as pd
import warnings
import wget
import sys
```

```python
pd.options.display.max_columns = None
```

```python
#create this bar_progress method which is invoked automatically from wget
def bar_progress(current, total, width=80):
  progress_message = "Downloading: %d%% [%d / %d] MBs" % (current / total * 100, current/1000000, total/1000000)
  # Don't use print() as it will print in new line every time.
  sys.stdout.write("\r" + progress_message)
  sys.stdout.flush()
```

```python
dados_municipio = pd.DataFrame()
for i in [2018, 2019, 2020, 2021,2022]:
   print(i)
   url = 'https://sage.saude.gov.br/dados/sisagua/vigilancia_parametros_basicos_' + str(i)+'.zip'
   filename = wget.download(url, bar=bar_progress)
   dados_ano = pd.read_csv(filename,
                         compression='zip', encoding='latin1', sep=';')
   filtro = (dados_ano['UF']=='RS')
   dados_rs_municipio = dados_ano[filtro].reset_index(drop=True)
   dados_municipio = pd.concat([dados_municipio, dados_rs_municipio])
   print(str(i) + ' - finalizado!')

dados_municipio

#dados_municipio.to_excel('dados_vigilancia_parametros_basicos_municipio_2014_2021.xlsx', index=False)
```

```python
dados_municipio['Parâmetro (parâmetros básicos)'].unique()
```

```python
colunas_manter = ['Regional de Saúde', 'Município',
       'Código IBGE','Código Forma de abastecimento',
       'Tipo da Forma de Abastecimento',
       'Nome da Forma de Abastecimento',
       'Sigla da Instituição', 'Nome da Instituição', 'Ano', 'Mês',
       'Data da Coleta','Parâmetro (parâmetros básicos)',
       'Resultado']

analises_manter = ['Fluoreto (mg/L)', 'Turbidez (uT)', 'Cloro residual livre (mg/L)',
       'Escherichia coli', ]

dados_municipio_limpo = dados_municipio[dados_municipio['Parâmetro (parâmetros básicos)'].isin(analises_manter)][colunas_manter]
dados_municipio_limpo.reset_index(drop=True, inplace=True)
```

```python
analises_manter = ['Fluoreto (mg/L)', 'Turbidez (uT)', 'Cloro residual livre (mg/L)',
       'Coliformes totais', 'Escherichia coli', ]

dados_fluoreto = dados_municipio_limpo[dados_municipio_limpo['Parâmetro (parâmetros básicos)']=='Fluoreto (mg/L)'].reset_index(drop=True)
dados_turbidez = dados_municipio_limpo[dados_municipio_limpo['Parâmetro (parâmetros básicos)']=='Turbidez (uT)'].reset_index(drop=True)
dados_cloro = dados_municipio_limpo[dados_municipio_limpo['Parâmetro (parâmetros básicos)']=='Cloro residual livre (mg/L)'].reset_index(drop=True)
dados_ecoli = dados_municipio_limpo[dados_municipio_limpo['Parâmetro (parâmetros básicos)']=='Escherichia coli'].reset_index(drop=True)

dados_fluoreto['Resultado'] = pd.to_numeric(dados_fluoreto['Resultado'].str.replace(',','.'))
dados_fluoreto['analise'] = ['Satisfatório'  if i <= 1.5 else 'Insatisfatório' for i in dados_fluoreto['Resultado']]

dados_turbidez['Resultado'] = pd.to_numeric(dados_turbidez['Resultado'].str.replace(',','.'))
dados_turbidez['analise'] = ['Satisfatório'  if i <= 5 else 'Insatisfatório' for i in dados_turbidez['Resultado']]

dados_cloro['Resultado'] = pd.to_numeric(dados_cloro['Resultado'].str.replace(',','.'))

analise = []
for i in dados_cloro['Resultado']:
  if i < 0.2:
    analise.append('Insatisfatório')

  elif i > 5:
    analise.append('Insatisfatório')

  else:
    analise.append('Satisfatório')

dados_cloro['analise'] = analise

dados_ecoli['analise'] = ['Satisfatório' if i == 'AUSENTE' else 'Insatisfatório' for i in dados_ecoli['Resultado']]

dados_2022 = pd.concat([dados_fluoreto, dados_turbidez, dados_cloro, dados_ecoli])
dados_2022.to_csv('dados_2022.csv', sep=';', index=False)
```

```python
dados_sac_ecoli = dados_2022[(dados_2022['Parâmetro (parâmetros básicos)']=='Escherichia coli')&(dados_2022['Tipo da Forma de Abastecimento']=='SAC')]
dados_sac_ecoli['chave'] = dados_sac_ecoli['Código Forma de abastecimento']+dados_sac_ecoli['Ano'].astype(str)
dados_sac_ecoli
```

```python
dados_fluoreto

analise = []
for resul,tipo in zip(dados_fluoreto['Resultado'],dados_fluoreto['Tipo da Forma de Abastecimento'] ):
  if tipo == 'SAA':
    if resul<=0.9 and resul >=0.6:
      analise.append('Satisfatório')
    else:
      analise.append('Insatisfatório')

  else:
    if resul<=1.5:
      analise.append('Satisfatório')
    else:
      analise.append('Insatisfatório')

dados_fluoreto['Analise_detalhe'] = analise

dados_fluoreto
```

```python
dados_fluoreto_empilhado = pd.melt(((pd.crosstab(dados_fluoreto['Analise_detalhe'], dados_fluoreto['Ano'], normalize='columns')*100).round(2)).reset_index(),
                                   id_vars='Analise_detalhe', value_vars=[2018,2019,2020,2021,2022], value_name='Porcentagem')
dados_fluoreto_empilhado
```

```python
import plotly.express as px


fig = px.bar(dados_fluoreto_empilhado.sort_values('Analise_detalhe', ascending=False), x="Ano", y="Porcentagem", color="Analise_detalhe",
             text='Porcentagem', color_discrete_sequence=["lightgreen", "indianred"])
fig.show()
```
