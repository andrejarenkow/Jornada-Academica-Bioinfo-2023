```python
import pandas as pd
import wget
from urllib.request import urlopen
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_url = 'https://ti.saude.rs.gov.br/covid19/download'
file_name = wget.download(file_url, 'dataset.csv')
```

```python
#casos = pd.read_csv('/content/dataset.csv', sep=';', encoding='latin-1')
casos = pd.read_csv(file_name, sep=';', encoding='utf-8')
casos['DATA_SINTOMAS'] = pd.to_datetime(casos['DATA_SINTOMAS'], dayfirst=True)
casos['DATA_CONFIRMACAO'] = pd.to_datetime(casos['DATA_CONFIRMACAO'], dayfirst=True)
casos['DATA_EVOLUCAO'] = pd.to_datetime(casos['DATA_EVOLUCAO'], dayfirst=True)
casos['DATA_EVOLUCAO_ESTIMADA'] = pd.to_datetime(casos['DATA_EVOLUCAO_ESTIMADA'], dayfirst=True)
casos['SEMANA'] = casos['DATA_SINTOMAS'].dt.week
casos.head()
```

```python
municipios = ['ALVORADA', 'CACHOEIRINHA', 'CAMPO BOM', 'CANOAS',
              'CAPÃO DA CANOA', 'ESTEIO', 'GRAVATAÍ', 'NOVO HAMBURGO',
              'PORTO ALEGRE', 'SAPUCAIA DO SUL', 'SÃO LEOPOLDO','TORRES','SANTA MARIA',
              'PASSO FUNDO', 'SANTA ROSA']

casos_municipios = casos[casos['MUNICIPIO'].isin(municipios)].reset_index(drop=True)
casos_municipios.head()
```

```python
tabela_dinamica = pd.pivot_table(data=casos_municipios, index='DATA_SINTOMAS', columns='MUNICIPIO', values='CRITERIO', aggfunc='count').fillna(0).reset_index()
tabela_dinamica
```

```python
tabela_dinamica_melt = pd.melt(tabela_dinamica, id_vars='DATA_SINTOMAS', value_vars=municipios)
tabela_dinamica_melt=tabela_dinamica_melt[tabela_dinamica_melt['DATA_SINTOMAS']>'2020-05-10']
tabela_dinamica_melt.columns = ['Data sintomas',	'Município',	'Casos']
tabela_dinamica_melt
```

```python
nome_planilha = 'Planilha coletas aberta'
sh = gc.open(nome_planilha)

worksheet = sh.worksheet("casos")

tabela_dinamica_melt['Data sintomas'] = tabela_dinamica_melt['Data sintomas'].astype(str)

worksheet.update([tabela_dinamica_melt.columns.values.tolist()] + tabela_dinamica_melt.values.tolist())
```

#plotando os casos junto com os dados em 2023 - Serraria e POA

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

tabela_dinamica_melt_poa = tabela_dinamica_melt[(tabela_dinamica_melt['Município']=='PORTO ALEGRE')&(tabela_dinamica_melt['Data sintomas']>='2023-01-01')]
carga_viral = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTZfjxdY8_x5WNd9_NE3QQPeche-dMdY5KdvNpq8H4W-lmUTidwrKpV0uLzLtihV7UAPIl68WvugMsN/pub?gid=0&single=true&output=csv')
carga_viral['Data de coleta'] = pd.to_datetime(carga_viral['Data de coleta'], dayfirst=True)

carga_viral = carga_viral[(carga_viral['Local de coleta']=='ETE Serraria')&(carga_viral['Data de coleta']>'2022-12-01')]
carga_viral['carga_viral_n1'] = pd.to_numeric(carga_viral['carga_viral_n1'])


fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(x=tabela_dinamica_melt_poa['Data sintomas'], y=tabela_dinamica_melt_poa['Casos'], name="Casos diários", mode="lines"),
    secondary_y=True
)

fig.add_trace(
    go.Bar(x=carga_viral['Data de coleta'], y=carga_viral['carga_viral_n1'], name="Carga Viral no esgoto"),
    secondary_y=False
)

fig.update_xaxes(title_text="Data")

# Set y-axes titles
fig.update_yaxes(title_text="Carga viral", secondary_y=False, range=[0,600000])
fig.update_yaxes(title_text="Casos diários", secondary_y=True, range=[0,250])

# Add figure title
fig.update_layout(
    title_text="Casos diários de COVID e Carga Viral de SARS-CoV-2 no Esgoto na ETE Serraria, Porto Alegre, 2023", template='none'
)

fig.show()
```
