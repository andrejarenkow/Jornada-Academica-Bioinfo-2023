# **Criando função que retorna, a partir do pdf, o que está em cada linha do arquivo pdf**

```python
def buscar_lines(arquivo_pdf):
  #criação de lista vazia
  lines = []
  #abrir o arquivo que está referenciado no caminho_pdf
  with pdfplumber.open(arquivo_pdf) as pdf:
  #coletar o número de páginas
      pages = pdf.pages
  #para cada página, extrai as linhas
      for page in pdf.pages:
          text = page.extract_text()
  #adiciona cada linha dentro da lista lines
          for line in text.split('\n'):
            lines.append(line)

  return lines
```
